```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicion√°rio de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
df = pd.DataFrame(dict_faturamento)

media_vendas = df['valor'].mean()

print(media_vendas)
```

```python
df['data_ref'] = pd.to_datetime(df['data_ref'])
```

```python
df = df.sort_values(by = 'data_ref')
df
```

```python
df['data_ref'] = pd.to_datetime(df['data_ref'])

df = df.sort_values(by='data_ref')

plt.figure(figsize=(12, 6))
plt.bar(df['data_ref'].dt.strftime('%Y-%m'), df['valor'], color='royalblue')

plt.xlabel("M√™s de Refer√™ncia", fontsize=12)
plt.ylabel("Valor (R$)", fontsize=12)
plt.title("Faturamento por M√™s", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['data_ref'].dt.strftime('%Y-%m'), df['valor'], marker='o', linestyle='-', color='royalblue', linewidth=2, markersize=8)


plt.xlabel("M√™s de Refer√™ncia", fontsize=12)
plt.ylabel("Valor (R$)", fontsize=12)
plt.title("Faturamento por M√™s", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```
